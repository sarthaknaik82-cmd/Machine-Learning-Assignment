```python
import pandas as pd

data = {
    "State": [
        "Uttar Pradesh", "Maharashtra", "Bihar", "West Bengal",
        "Tamil Nadu", "Karnataka", "Gujarat", "Rajasthan",
        "Madhya Pradesh", "Kerala"
    ],
    "Votes_Lakhs": [912, 590, 480, 510, 430, 390, 410, 360, 370, 280]
}

df = pd.DataFrame(data)
df

##Line plot
import matplotlib.pyplot as plt
plt.figure()
plt.plot(df["State"], df["Votes_Lakhs"], marker='o')
plt.xticks(rotation=45)
plt.xlabel("State")
plt.ylabel("Votes (in Lakhs)")
plt.title("Votes of Indian States")
plt.show()

##Pie plot
plt.figure()
plt.pie(df["Votes_Lakhs"], labels=df["State"], autopct='%1.1f%%')
plt.title("Vote Share by the Indian States")
plt.show()

##bar plot
plt.figure()
plt.hist(df["Votes_Lakhs"], bins=5)
plt.xlabel("Votes (in Lakhs)")
plt.ylabel("Frequency")
plt.title("Distribution of Votes Across States")
plt.show()

import seaborn as sns
##sns barplot
sns.barplot(x="Votes_Lakhs", y="State", data=df)
plt.title("Votes by State ")
plt.show()

##boxplot
sns.boxplot(x=df["Votes_Lakhs"])
plt.title("Vote Distribution Boxplot")
plt.show()
```
