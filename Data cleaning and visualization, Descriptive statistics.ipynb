# Task
Perform a statistical analysis of the numerical columns in the `df` DataFrame, including descriptive statistics (mean, median, standard deviation), mode, and a correlation matrix.

## Identify Numerical Columns

### Subtask:
Examine the DataFrame to identify all numerical columns suitable for statistical analysis.


**Reasoning**:
To identify numerical columns, I will display the data types of all columns in the DataFrame using `df.dtypes`.



```python
print(df.dtypes)
```

**Reasoning**:
Based on the output of `df.dtypes`, all columns except 'animal_name' are integers (`int64`), which are numerical. I will now create a list of these numerical column names.



```python
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()
print("Numerical columns identified:", numerical_cols)
```

```python
print(df.head())
```

## Calculate Descriptive Statistics

### Subtask:
Compute the mean, median, and standard deviation for all identified numerical columns.


**Reasoning**:
To compute the mean, median, and standard deviation for the identified numerical columns, I will select these columns from the DataFrame and then use built-in pandas functions to calculate the required statistics. The `.describe()` method provides mean and standard deviation, and the `.median()` method will provide the median.



```python
numerical_df = df[numerical_cols]

desc_stats = numerical_df.head(5).describe().loc[['mean']]
median_stats = numerical_df.head(5).median().to_frame(name='median').T


print("Descriptive Statistics (Mean ) for first 5 rows:")
print(desc_stats)
print("\nMedian for first 5 rows:")
print(median_stats)
```

```python
numerical_df = df[numerical_cols]
desc_stats = numerical_df.head(5).describe().loc[['std']]
print("Descriptive Statistics (std ) for first 5 rows:")
print(desc_stats)
```

## Calculate Mode

### Subtask:
Determine the mode for each numerical column, as this is not always covered by standard descriptive statistics functions.


**Reasoning**:
To determine the mode for each numerical column, I will apply the `.mode()` method to the `numerical_df` DataFrame. Since a column can have multiple modes, I will select the first mode using `.iloc[0]` to get a single representative value for each column.



```python
mode_stats = numerical_df.head(5).mode().iloc[0]
print("Mode for each numerical column (first 5 rows):")
print(mode_stats)
```

## Calculate Correlation Matrix

### Subtask:
Compute the correlation matrix for all identified numerical columns to understand the relationships between them.

**Reasoning**:
To compute the correlation matrix, I will apply the `.corr()` method to the `numerical_df` DataFrame, which contains all the identified numerical columns.



```python
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:")
print(correlation_matrix)
```

# Task
Generate a heatmap of the correlation matrix to visualize the linear relationships between all numerical features.

## Visualize Correlation Matrix

### Subtask:
Generate a heatmap of the correlation matrix to visually represent the linear relationships between all numerical features.


**Reasoning**:
To visualize the correlation matrix, I will generate a heatmap using `seaborn.heatmap()`. I will set an appropriate figure size, add annotations for readability, and include a title for better understanding.



```python
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Distribution of 'legs'

### Subtask:
Create a histogram or count plot to visualize the frequency distribution of the 'legs' feature, showing the most common number of legs among the animals.


**Reasoning**:
To visualize the frequency distribution of the 'legs' feature and show the most common number of legs, I will create a count plot using seaborn. This will effectively display the counts of each unique value in the 'legs' column.



```python
plt.figure(figsize=(8, 6))
sns.countplot(x='legs', data=df, palette='viridis', hue='legs', legend=False)
plt.xlabel('Number of Legs')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Legs Among Animals')
plt.show()
```

## Counts of 'hair' and 'feathers'

### Subtask:
Generate bar plots (or count plots) to show the distribution of animals with and without 'hair' and 'feathers', as these are binary categorical features.


**Reasoning**:
To visualize the distribution of 'hair' and 'feathers' features, I will create two count plots side-by-side using `plt.subplots` and `seaborn.countplot`, as instructed. This will clearly show the counts of animals with and without these characteristics.



```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(x='hair', data=df, ax=axs[0], palette='viridis', hue='hair', legend=False)
axs[0].set_xlabel('Hair')
axs[0].set_ylabel('Count')
axs[0].set_title('Distribution of Animals by Hair')

sns.countplot(x='feathers', data=df, ax=axs[1], palette='viridis', hue='feathers', legend=False)
axs[1].set_xlabel('Feathers')
axs[1].set_ylabel('Count')
axs[1].set_title('Distribution of Animals by Feathers')

plt.tight_layout()
plt.show()
```

## Counts of 'milk' and 'eggs'

### Subtask:
Generate bar plots (or count plots) to show the distribution of animals producing 'milk' and laying 'eggs', exploring these key biological characteristics.


**Reasoning**:
To visualize the distribution of 'milk' and 'eggs' features as requested, I will create two count plots side-by-side using `plt.subplots` and `seaborn.countplot`, applying the specified labels, titles, and plot settings.



```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

sns.countplot(x='milk', data=df, ax=axs[0], palette='viridis', hue='milk', legend=False)
axs[0].set_xlabel('Milk Production (0=No, 1=Yes)')
axs[0].set_ylabel('Count')
axs[0].set_title('Distribution of Animals by Milk Production')

sns.countplot(x='eggs', data=df, ax=axs[1], palette='viridis', hue='eggs', legend=False)
axs[1].set_xlabel('Egg Laying (0=No, 1=Yes)')
axs[1].set_ylabel('Count')
axs[1].set_title('Distribution of Animals by Egg Laying')

plt.tight_layout()
plt.show()
```

## Relationship between 'aquatic' and 'fins'

### Subtask:
Create a visualization (e.g., a grouped bar chart or heatmap) to explore the relationship between 'aquatic' animals and the presence of 'fins', highlighting how often these features co-occur.


**Reasoning**:
To visualize the relationship between 'aquatic' animals and the presence of 'fins', I will create a grouped bar chart using `seaborn.countplot`, as instructed, setting 'aquatic' on the x-axis and 'fins' as the hue to show their co-occurrence. I will then set appropriate labels, titles, and legend for clarity.



```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x='aquatic', hue='fins', data=df, palette='viridis', ax=ax, legend=True)
ax.set_xlabel('Aquatic (0=No, 1=Yes)')
ax.set_ylabel('Count')
ax.set_title('Relationship Between Aquatic Animals and Presence of Fins')

handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=['No Fins', 'Has Fins'], title='Fins (0=No, 1=Yes)')

plt.tight_layout()
plt.show()
```
