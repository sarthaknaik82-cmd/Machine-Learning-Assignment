## 4. Create a tuple and try to change a value. What happens?

```python
my_tuple = (1, 2, 3)
try:
    my_tuple[0] = 100
except TypeError as e:
    print(f"Error: {e}")
```

As shown above, trying to change a value in a tuple results in a `TypeError`. This is because tuples are **immutable**, meaning their elements cannot be changed, added, or removed after the tuple's creation.

## 13. Reverse a list.

```python
my_list = [1, 2, 3, 4, 5]
my_list.reverse() # Reverses the list in-place
print(f"In-place reversed list: {my_list}")

# Alternative: create a new reversed list without modifying the original
original_list = [6, 7, 8, 9, 10]
new_reversed_list = original_list[::-1]
print(f"New reversed list: {new_reversed_list}")
print(f"Original list remains: {original_list}")
```

## 14. Sort a list.

```python
my_list = [5, 2, 8, 1, 9]
my_list.sort() # Sorts the list in-place
print(f"In-place sorted list: {my_list}")

# Alternative: create a new sorted list without modifying the original
unsorted_list = [10, 3, 7, 6, 4]
new_sorted_list = sorted(unsorted_list)
print(f"New sorted list: {new_sorted_list}")
print(f"Original list remains: {unsorted_list}")
```

## 17. Show that set does not allow duplicates.

```python
my_set = {1, 2, 3}
print(f"Initial set: {my_set}")
my_set.add(2) # Attempt to add an existing element
my_set.add(5) # Add a new element
print(f"Set after adding 2 (again) and 5: {my_set}")
```

As you can see, even after trying to add `2` again, the set `{1, 2, 3}` remains `{1, 2, 3}` (or `{1, 2, 3, 5}` when `5` is added), demonstrating that sets only store unique elements. Duplicates are automatically ignored.

## 20. Count vowels in a string.

```python
my_string = "Hello World! This is a Test String."
vowels = "aeiouAEIOU"
vowel_count = 0
for char in my_string:
    if char in vowels:
        vowel_count += 1
print(f"The number of vowels in '{my_string}' is: {vowel_count}")
```
